```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
%matplotlib notebook
```

```python
# Load the thermodynamic state space
Varray = np.loadtxt('.V.txt') # In m^3
Tarray = np.loadtxt('.T.txt') # In K

# Load the old state functions & transpose (In kJ/mol ... rows are for volume, columns are for temperature)
Ugrid1 = np.loadtxt('.Ugrid1.txt').T
Ugrid2 = np.loadtxt('.Ugrid2.txt').T
Ugrid3 = np.loadtxt('.Ugrid3.txt').T

# Make new state space grid
Vgrid, Tgrid = np.meshgrid(Varray,Tarray)
```

```python

# Open up a 3d figure window
ax = plt.figure().gca(projection='3d') # Set up a three dimensional graphics window 

# Plot surface 1
ax.plot_surface(Tgrid, Vgrid*1e3, Ugrid1, color='blue') # Make the mesh plot
ax.set_ylabel('V (L)') # Label axes
ax.set_xlabel('T (K)')
ax.set_zlabel('U (kJ/mol)')

```

```python
# Open up a 3d figure window
ax = plt.figure().gca(projection='3d') # Set up a three dimensional graphics window 

# Prepping the axis labels
xlabel = "T "
ylabel = "V "
zlabel = "T "

# Graph the pressure
ax.plot_surface(Tgrid, Vgrid, Tgrid, color='teal') # Make the mesh plot T(T,V)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.set_zlabel(zlabel)

# Open up a 3d figure window
ax = plt.figure().gca(projection='3d') # Set up a three dimensional graphics window 

# Prepping the axis labels
xlabel = "T "
ylabel = "V "
zlabel = "V "

# Graph the pressure
ax.plot_surface(Tgrid, Vgrid, Vgrid, color='violet') # Make the mesh plot V(T,V)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.set_zlabel(zlabel)
```
